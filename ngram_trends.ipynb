```python
import csv 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib.font_manager import FontProperties
from scipy import statsmir

results_df = pd.read_csv("TBU")
print(results_df.columns)
# Set the style and font
plt.style.use('default')  # Using default style instead of seaborn
font = FontProperties(family='Times New Roman')
plt.rcParams['font.family'] = 'Times New Roman'
# Set global font size to 11
plt.rcParams['font.size'] = 14
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['xtick.labelsize'] = 14
plt.rcParams['ytick.labelsize'] = 14
plt.rcParams['legend.fontsize'] = 14

# Create a wide figure
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(12, 3))

# Plot 1: Mono_Pct vs Hallucination with regression line and confidence interval
sns.scatterplot(data=results_df, 
            x='Mono_Pct', 
            y='Hallucination',
            ax=ax1,
            color = "lightblue", 
            edgecolor = "blue")
ax1.set_xlabel('Monofact Rate (%)', fontproperties=font, fontsize = 14)
ax1.set_ylabel('Hallucination Rate (%)', fontproperties=font, fontsize = 14)
ax1.set_ylim(0, 0.5)  # Set y-axis limits
ax1.set_yticks(np.arange(0, 0.6, 0.1))  # Set tick marks every 0.1
ax1.set_xlim(0., 1)  # Set y-axis limits
ax1.set_xticks(np.arange(0, 1.01, 0.25))  # Set tick marks every 0.1
ax1.grid(True, linestyle='--', alpha=0.7)

sns.scatterplot(data=results_df,
                x='Pareto',
                y='Mono_Pct',
                ax=ax2,
                alpha=0.6,
            color = "lightblue", 
            edgecolor = "blue")
# ax2.plot(x_poly, p(x_poly), 'r-', linewidth=1.5)
# ax2.set_ylim(0, 0.9)  # Set y-axis limits
ax2.set_yticks(np.arange(0, 1.001, 0.25))  # Set tick marks every 0.1
ax2.set_xticks(np.arange(1, 4.001, 0.5))
ax2.set_xlabel('Shape Parameter ()', fontproperties=font, fontsize = 14)
ax2.set_ylabel('Monofact Rate (%)', fontproperties=font, fontsize = 14)
ax2.grid(True, linestyle='--', alpha=0.7)


sns.scatterplot(data=results_df,
                x='Miscalibration',
                y='Mono_Pct',
                alpha=0.6,
                color='gray',
                edgecolor = "black",
                ax=ax3)

ax3.set_xlabel('Miscalibration (%)', fontproperties=font, fontsize = 14)
ax3.set_ylabel('Monofact Rate (%)', fontproperties=font, fontsize = 14)
ax3.set_ylim(0, 1.0000001)  # Set y-axis limits
ax3.set_yticks(np.arange(0, 1.000000001, 0.25))  # Set tick marks every 0.1
ax3.set_xlim(0, 0.750000001)  # Set y-axis limits
ax3.set_xticks(np.arange(0, 0.750001, 0.25))  # Set tick marks every 0.1
ax3.grid(True, linestyle='--', alpha=0.7)

sns.scatterplot(data=results_df,
                x='Miscalibration',
                y='CE_reg',
                alpha=0.6,
                color='gray',
                edgecolor = "black",
                ax=ax4)

ax4.set_yticks(np.arange(0, 0.75000001, 0.25))  # Set tick marks every 0.1
ax4.set_xticks(np.arange(0,1, 0.25))
ax4.set_xlabel('Emp. KL Div.', fontproperties=font, fontsize = 14)
ax4.set_ylabel('Miscalibration (%)', fontproperties=font, fontsize = 14)
ax4.grid(True, linestyle='--', alpha=0.7)


# Adjust layout
plt.tight_layout(pad=1.5) 

# Add extra space for caption
plt.subplots_adjust(bottom=0.35)

plt.savefig('TBU', 
            dpi=600, 
            bbox_inches='tight',
            pad_inches= 0.05)
plt.show()
```
