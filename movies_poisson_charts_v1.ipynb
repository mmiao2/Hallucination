```python
import csv 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib.font_manager import FontProperties
from scipy import stats

results_df = pd.read_csv("path")

# Set the style and font
plt.style.use('default')  # Using default style instead of seaborn
font = FontProperties(family='Times New Roman')
plt.rcParams['font.family'] = 'Times New Roman'

# Set global font size to 11
plt.rcParams['font.size'] = 11
plt.rcParams['axes.labelsize'] = 11
plt.rcParams['axes.titlesize'] = 11
plt.rcParams['xtick.labelsize'] = 11
plt.rcParams['ytick.labelsize'] = 11
plt.rcParams['legend.fontsize'] = 11

# Create a wide figure
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 3))

sns.scatterplot(data=results_df, 
            x='Mono_Pct', 
            y='Hallucination',
            ax=ax1,
            color = "blue")
ax1.set_xlabel('Mono-fact Rate', fontproperties=font)
ax1.set_ylabel('Hallucination Rate', fontproperties=font)

ax1.grid(True, linestyle='--', alpha=0.7)

sns.scatterplot(data=results_df,
                x='Lambda',
                y='Mono_Pct',
                ax=ax2,
                alpha=0.6,
                color='blue')
ax2.set_xlabel('Lambda (Î»)', fontproperties=font)
ax2.set_ylabel('Monofact Rate', fontproperties=font)
ax2.grid(True, linestyle='--', alpha=0.7)


# Adjust layout
plt.tight_layout(pad=1.5) 
plt.subplots_adjust(bottom=0.35)

plt.savefig('path', 
            dpi=300, 
            bbox_inches='tight',
            pad_inches=0.2)
plt.show()
```
